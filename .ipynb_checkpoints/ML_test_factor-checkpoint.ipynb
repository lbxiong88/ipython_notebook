```python
codeall=ps.ExcelFile(u'r:\\成份股市值更新.xlsx').parse(u'全市场市值更新',header=1,index_col=0).dropna()
```

```python
# 板块指标股 #常用板块
# u'上证50指数成份',u'沪深300',u'中证500',u'全部A股(非ST)',u'融资融券标的',
# u'CS银行'，u'CS非银行金融'
wret=w.wset("SectorConstituent","date=20150831","sector=%s"%u'CS非银行金融')
codesubretdf=ps.DataFrame(wret.Data,index=wret.Fields).T.set_index("date")
codesubretdf.index=codesubretdf.index-ps.datetools.Milli(5)
codesubretdf
```

```python
#历史PE_TTM数据
wret=w.wsd(list(codesubretdf.wind_code),'pe_ttm','2010-1-1','2015-8-31','Period=M','Fill=Previous')
pettmretdf=ps.DataFrame(wret.Data,columns=(ps.DatetimeIndex(wret.Times)-ps.datetools.Milli(5)),index=wret.Codes).T;pettmretdf
```

```python
#自由市场换手率数据
wret=w.wsd(list(codesubretdf.wind_code),'free_turn','2010-1-1','2015-8-31','Period=M','Fill=Previous')
factorretdf=ps.DataFrame(wret.Data,columns=(ps.DatetimeIndex(wret.Times)-ps.datetools.Milli(5)),index=wret.Codes).T;factorretdf
```

```python
#收益率
wret=w.wsd(list(codesubretdf.wind_code),'pct_chg','2010-1-1','2015-8-31','Period=M','Fill=Previous')
pct_chgretdf=ps.DataFrame(wret.Data,columns=(ps.DatetimeIndex(wret.Times)-ps.datetools.Milli(5)),index=wret.Codes).T;pct_chgretdf
```

```python
qcutretdic={}
for k,v in mktcapretdf.T.iteritems():
    qcutretdic[k]=ps.qcut(v,10,labels=list('ABCDEFGHIJ')).astype('string')
qcutrank=ps.DataFrame(qcutretdic).T;mktrank=qcutrank.copy()
mktranktrans=ps.melt(mktrank.reset_index(),id_vars='index')
mktranktrans['mkt_trans']=mktranktrans.value.map(ps.Series(np.linspace(0,1,10),index=list('ABCDEFGHIJ')));mktranktrans
```

```python
qcutretdic={}
for k,v in factorretdf.T.iteritems():
    qcutretdic[k]=ps.qcut(v,10,labels=list('ABCDEFGHIJ')).astype('string')
qcutrank=ps.DataFrame(qcutretdic).T;factorrank=qcutrank.copy()
factorranktrans=ps.melt(factorrank.reset_index(),id_vars='index')
factorranktrans['factorrank_trans']=factorranktrans.value.map(ps.Series(np.linspace(0,1,10),index=list('ABCDEFGHIJ')));factorranktrans
```

```python
qcutretdic={}
for k,v in pct_chgretdf.T.iteritems():
    qcutretdic[k]=ps.qcut(v,5,labels=list('ABCDE')).astype('string')
qcutrank=ps.DataFrame(qcutretdic).T;pct_chgrank=qcutrank.copy()
pct_chgranktrans=ps.melt(pct_chgrank.reset_index(),id_vars='index');
pct_chgranktrans['pctrank_trans']=pct_chgranktrans.value.map(ps.Series(np.linspace(0,1,10),index=list('ABCDEFGHIJ')));pct_chgranktrans
```

```python
qcutretdic={}
for k,v in pettmretdf.T.iteritems():
    qcutretdic[k]=ps.qcut(v,10,labels=list('ABCDEFGHIJ')).astype('string')
qcutrank=ps.DataFrame(qcutretdic).T;perank=qcutrank.copy()
peranktrans=ps.melt(perank.reset_index(),id_vars='index')
peranktrans['pe_trans']=peranktrans.value.map(ps.Series(np.linspace(0,1,10),index=list('ABCDEFGHIJ')));peranktrans
```

```python
factor_all=ps.merge(peranktrans,
                    ps.merge(pct_chgranktrans,
                        ps.merge(mktranktrans,factorranktrans,on=['index','variable']),
                        on=['index','variable']),
                            on=['index','variable'])
factor_all
```

```python
factor_all['month']=[x.month for x in factor_all['index']]
factor_all=factor_all.drop([u'value',u'value_x',u'value_y'],axis=1)
factor_all_ind=ps.merge(factor_all,codeall[['ind1']],how='left',left_on='variable',right_index=True)
factor_all_ind['ind']=factor_all_ind.ind1.map(ps.Series({item:i for i,item in enumerate(codeall.ind1.unique())}));factor_all_ind
```

```python
index_bench='000300.SH'
wret=w.wsd(index_bench,'pct_chg','2010-1-1','2015-8-31','Period=M','Fill=Previous')
hs300_chgretdf=ps.DataFrame(wret.Data,columns=(ps.DatetimeIndex(wret.Times)-ps.datetools.Milli(5)),index=wret.Codes).T;hs300_chgretdf
```

```python
targetdf=ps.merge(ps.melt(pct_chgretdf.shift(-1).reset_index(),id_vars='index'),hs300_chgretdf.shift(-1),left_on='index',right_index=True)
targetdf['target']=where(targetdf['value']>(targetdf[index_bench]+0.2),1,0);targetdf
```

```python
trainset=ps.merge(targetdf.dropna(),factor_all_ind,on=['index','variable'])[['pctrank_trans',
                                                                             'mkt_trans','factorrank_trans','month','ind','pe_trans','target','index','variable','value']].dropna()
trainset=trainset[-trainset['index'].isin([])];trainset
```

```python
trainset.groupby('index')['target'].sum()
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
#0:pctrank_trans,
#         1:mkt_trans,
#             2:factorrank_trans,
#                 3:month
#                         4:ind
#                                5:pe_ttm
features=[0,1,2,5]
trainsetsel_date =  trainset.ix[:]
print trainsetsel_date.irow(0)['index'],trainsetsel_date.irow(-1)['index']
trainsetsel= trainsetsel_date.ix[:,features].values

from sklearn.cross_validation import train_test_split
features_train, features_test, target_train, target_test = train_test_split(
    trainsetsel, trainsetsel_date.ix[:,6].values, test_size=0.05, random_state=randint(100))
features_train.shape,features_test.shape,target_train.shape,target_test.shape

hstrain=RandomForestClassifier(max_depth=3,n_estimators=100)
hstrain.fit(features_train, target_train)

trainset['prodict_ret']=hstrain.predict(trainset.ix[:,features])
prodict_retdf=trainset.pivot(index='index',columns='variable',values='prodict_ret')
print hstrain.score(trainset.ix[:,features], trainset['target'])
# print hstrain.feature_importances_
print 'Long/short',(pct_chgretdf[prodict_retdf.shift(1)==1].mean(1)-(pct_chgretdf[prodict_retdf.shift(1)==0].mean(1))).mean()
print 'Alpha',(pct_chgretdf[prodict_retdf.shift(1)==1].mean(1)-hs300_chgretdf[index_bench].mean()).mean()
hstrain.feature_importances_
```
